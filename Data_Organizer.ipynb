```python
import pandas as pd
```

```python
consumer_data=pd.read_csv('consumer_data.csv')
```

```python
for col in range(len(demo_cols)):
    print demo_cols.iloc[col][0], demo_cols.iloc[col][1]
    df.rename(columns={demo_cols.iloc[col][0]: demo_cols.iloc[col][1]}, inplace=True)
```

```python
df
```

```python
new_df=pd.DataFrame()
```

```python
new_df
```

```python
for col in demo_cols.iloc[:,1].unique():
    try:
        new_df[col]=df[col].sum(axis=1)
    except:
        new_df[col]=df[col]

        
```

```python
#new_df=pd.DataFrame()
for col in ['U','N','Alcohol','EO','S']:
    new_df[col]=df[col].sum(axis=1)
```

```python
new_df
```

```python
latest_df=pd.DataFrame()
latest_df=new_df.copy(deep=True)
```

```python
latest_df
```

```python
latest_df['tot_expenditure']=0
for col in ['U','N','Alcohol','EO','S']:
    latest_df['tot_expenditure']+= latest_df[col]
```

```python
latest_df
```

```python
for col in latest_df.columns:
    if col[0]=='I':
        latest_df[col+'_rate']=(latest_df[col]/latest_df['HH'])*100
    elif col[0]=='P' and (col != 'Pop' or col[:2] != 'PF' or col[:2] != 'PM'):
        latest_df[col+'_rate']=(latest_df[col]/latest_df['Pop'])*100
    elif col[:2]=='PF':
        latest_df[col+'_rate']=(latest_df[col]/latest_df['Pop_F'])*100
    elif col[:2]=='PM':
        latest_df[col+'_rate']=(latest_df[col]/latest_df['Pop_M'])*100
    elif col in ['U','N','Alcohol','EO','S']:
        latest_df[col+'_exp_rate']=(latest_df[col]/latest_df['tot_expenditure'])*100
        latest_df[col+'_rate_HH']=(latest_df[col]/latest_df['HH'])
        latest_df[col+'_rate_Pop']=(latest_df[col]/latest_df['Pop'])
```

```python
final_data_df=latest_df.copy(deep=True)
```

```python
final_data_df
```

```python
final_data_df=final_data_df.reset_index()
```

```python
final_data_df.to_csv('exp_n_demo.csv')
```
